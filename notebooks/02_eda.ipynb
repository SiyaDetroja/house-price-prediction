# Exploratory Data Analysis (EDA)

## Objective
The objective of this notebook is to explore the dataset visually and statistically to identify patterns, relationships, and potential issues before model building.


```python
import matplotlib.pyplot as plt 
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
file_path = '/content/drive/MyDrive/House Price India.csv'
data = pd.read_csv(file_path)

```

```python
plt.figure(figsize = (10,6))
sns.histplot(data['Price'], bins = 20)
plt.xlabel('Price')
plt.ylabel("Frequency")
plt.title("Distribution of House Prices")
plt.show()
```

### Price Distribution Analysis

- The distribution of house prices is right-skewed.
- Most houses are concentrated in the lower price range.
- A small number of houses have very high prices, creating a long right tail.
- This indicates the presence of high-value outliers.


```python
plt.figure(figsize=(8,4))
sns.scatterplot(x='house area(excluding basement)', y='Price', data=data)
plt.xlabel('Area (sq ft)')
plt.ylabel('Price (INR)')
plt.title('Area vs Price')
plt.show()
```

### Area vs Price Analysis

- There is a positive relationship between house area and price.
- As area increases, house prices generally increase.
- The relationship is not perfectly linear, indicating that other features also influence price.
- Price variability increases with area, and a few outliers are present.


```python
plt.figure()
plt.scatter(data["No of bedrooms"], data["Price"])
plt.xlabel("Bedrooms")
plt.ylabel("Price")
plt.title("Bedrooms vs Price")
plt.show()
```

```python
plt.figure(figsize = (10,6))
plt.scatter(data['waterfront present'],data['Price'])
plt.xlabel('waterfront present')
plt.ylabel('price')
plt.title('waterfront present vs price')
plt.show()
```

- Larger houses tend to have higher prices.
- The relationship is not perfectly linear, suggesting non-linear models may perform better.


```python
for col in data.columns:
    plt.figure()
    sns.boxplot(x=data[col])
    plt.title(f"Box Plot of {col}")
    plt.show()
```

### Box Plot Analysis

- Box plots were used to detect outliers in numerical features.
- Several features show the presence of outliers.
- The target variable (Price) contains high-value outliers, which is consistent with the skewed distribution observed earlier.
- Outliers will be handled carefully during preprocessing.


```python
plt.figure(figsize=(18, 10))
sns.heatmap(data.corr(),annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

- Some features show strong correlation with house price.
- Highly correlated features may be important predictors.
